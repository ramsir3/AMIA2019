```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import normalize, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
import sklearn.metrics as skm
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('~/Projects/Research/data/processed/HOUR_00001.csv', na_values='?')
```

```python
data.head()
```

```python
meanImputeData = data.fillna(data.mean())
meanImputeData.head()
```

```python
# comment this section out. I have since adjusted the script to sort items in the correct order
cols = meanImputeData.columns.tolist()
cols = cols[:2] + [cols[4]] + cols[2:4] + cols[5:]
meanImputeData = meanImputeData[cols]
meanImputeData.head()
```

```python
print(meanImputeData.isna().sum())
```

```python
meanImputeDataArray = meanImputeData.values[:, 3:]
x = meanImputeDataArray[:, :-6]
y = meanImputeDataArray[:, -1]
x = normalize(x, axis=0)
ohe = OneHotEncoder()
ohe.fit(y.reshape(-1, 1))
print(x.shape)
print(y.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
print(x_train.shape)
print(y_train.shape)
ohe_y_train = ohe.transform(y_train.reshape(-1,1))
ohe_y_test = ohe.transform(y_test.reshape(-1,1))
```

```python
def runModel(model, x, y, x_test, y_test):
    m = OneVsRestClassifier(model)
    m.fit(x,y)
    y_predict = m.predict(x_test)
    
    return y_predict
```

```python
def runMetrics(y_test, y_predict, metrics):
    # ohe_yp = np.zeros_like(y_predict)
    # ohe_yp[np.arange(len(y_predict)), y_predict.argmax(1)] = 1
    # ohe_yp = ohe.reverse_transform(b)
    out = []
    for m, kwargs in metrics:
        print(m.__name__)
        out.append(m(y_test, y_predict, **kwargs))
```

```python
def multiclass_auc(y_test, y_score):
    print(y_test.shape)
    print(y_score.shape)
    n_classes = 4
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = skm.roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = skm.auc(fpr[i], tpr[i])
    return sum(roc_auc.values())/n_classes
```

```python
metrics = [
    (skm.classification_report, {}),
    (skm.accuracy_score, {}),
#     (multiclass_auc, {}),
]
```

```python
models = {
    'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'GaussianProcessClassifier': GaussianProcessClassifier(),
    'SGDClassifier': SGDClassifier(),
    'GaussianNB': GaussianNB(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'MLPClassifier': MLPClassifier(),
}
```

```python
with open("output.csv", 'a') as fout:
    for n, m in models.items():
        print(n)
        y_predict = runModel(m, x_train, y_train, x_test, y_test)
        scores = runMetrics(y_test, y_predict, metrics)
        fout.write(n+'\n')
        for s in scores:
            fout.write(str(s)+'\n')
```
