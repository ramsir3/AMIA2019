```python
from rnn_utils import *
from constants import RAW_PATH, DATA_PATH
import os
import pandas as pd
```

```python
idsPath = os.path.join(RAW_PATH, 'd_ids_split.pickle')
datapath = os.path.join(DATA_PATH, 'bypt500')
```

```python
split_ids = getSplitIds(idsPath)
db = DataBatch(datapath, split_ids, batchSize=10)
print([(k, len(v)) for k, v in db.files.items()])
```

```python
trainBatches = db.getBatchIterator('devel')
```

```python
header = db.getHeaders()
```

```python
for b, batch in enumerate(trainBatches):
    v, h, f = batch.shape
    hdr = {h: i for i, h in enumerate(header)}
    s = hdr['STAGE48']

    splot = batch[:,:,s]

    df = pd.DataFrame(splot.T, columns=range(v), index=range(h))
    df.plot()
    
    if b == 10:
        break
```

```python
batch = None
for b, batch in enumerate(trainBatches):  
    if b == 4:
        break

v, h, f = batch.shape
hdr = {h: i for i, h in enumerate(header)}
s = hdr['STAGE48']

splot = batch[:,:,s]

df = pd.DataFrame(splot.T, columns=range(v), index=range(h))
df.to_csv('examplePatients.csv')
```
